```python
import pickle
import sympy as sp
```

Bivalent

```python
biv_lst = []
for i in range(107):
    with open("m%d/res_0.out" % i,'rb') as inf:
        res_0_file = pickle.load(inf)
    print(i,sp.factor(sum([res_0_file[i][0] for i in range(len(res_0_file))])))
    biv_lst.append([i,sp.factor(sum([res_0_file[i][0] for i in range(len(res_0_file))]))])
```

Quad A

```python
quad_A_lst = []
for i in range(107):
    with open("m%d/res_1_0.out" % i,'rb') as inf:
        res_0_file = pickle.load(inf)
    print(i,sp.factor(sum([res_0_file[i][0] for i in range(len(res_0_file))])))
    quad_A_lst.append([i,sp.factor(sum([res_0_file[i][0] for i in range(len(res_0_file))]))])
```

Quad B

```python
quad_B_lst = []
for i in range(107):
    with open("m%d/res_1_1.out" % i,'rb') as inf:
        res_0_file = pickle.load(inf)
    print(i,sp.factor(sum([res_0_file[i][0] for i in range(len(res_0_file))])))
    quad_B_lst.append([i,sp.factor(sum([res_0_file[i][0] for i in range(len(res_0_file))]))])
```

Quad C

```python
quad_C_lst = []
for i in range(107):
    with open("m%d/res_1_2.out" % i,'rb') as inf:
        res_0_file = pickle.load(inf)
    print(i,sp.factor(sum([res_0_file[i][0] for i in range(len(res_0_file))])))
    quad_C_lst.append([i,sp.factor(sum([res_0_file[i][0] for i in range(len(res_0_file))]))])
```

```python
comb_lst = []
for i in range(len(biv_lst)):
    comb_lst.append([biv_lst[i][0],sp.simplify(biv_lst[i][1] + quad_A_lst[i][1] + quad_B_lst[i][1] + quad_C_lst[i][1])])
```
